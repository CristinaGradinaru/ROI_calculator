```python
class homeownerCashFlow:
    def __init__(self):
        
        choice = input('Did you pay for your house in full? ')
        if choice == 'yes':
            self.house_value = int(input('How much does your house cost? '))
            self.totalMonthlyIncome()
        elif choice == 'no':
            print('Let\'s find out your total monthly income. ')
        else:
            print(f'{choice} is not a valid answer.')
            self.totalMonthlyIncome()
    
                  
    def totalMonthlyIncome(self):
        
        self.rent = int(input("What is your monthly rental income?"))
        self.other_inc = int(input("Do you have other monthly income (ex. laundry, storage, etc.)? If \'yes\' type the amount, otherwise type in \'0\'"))
        self.monthly_income = self.rent + self.other_inc
        print(f'Your monthly income is {self.monthly_income} Now let\'s find out your total monthly expenses! ')
        self.totalMonthlyExpenses()
        
    def totalMonthlyExpenses(self, monthly_expenses=0):
        self.monthly_expenses= monthly_expenses
        
        self.taxes = int(input('How much are your monthly property taxes expenses? '))
        monthly_expenses += self.taxes
        self.insurance = int(input('How much are your monthly insurance expenses? '))
        monthly_expenses += self.insurance
        self.utilities = int(input('Do you have utility expenses? If yes how much are your monthly utilities expenses(type 0 for none)? '))
        monthly_expenses += self.utilities
        self.hoa = int(input('How much are your monthly home owner\'s association expenses? '))
        monthly_expenses += self.hoa
        self.lawn_snow = int(input('How much are your monthly lawn maintenance expenses (type 0 for None)? '))
        monthly_expenses += self.lawn_snow
        self.vacancy = int(input('How much do you want to set aside for vacancy expenses? (5% from rental income is recommended)'))
        monthly_expenses += self.vacancy
        self.repairs = int(input('How much do you want to set aside for repairs & CapEx  expenses? (5% from rental income is recommended)'))
        monthly_expenses += self.repairs
        self.property_management = int(input('How much are your property management expenses? '))
        monthly_expenses += self.property_management
        self.mortgage = int(input('How much do you pay for your mortgage per month (type 0 if None)? '))
        monthly_expenses += self.mortgage
        print(f"Your total monthly expenses are {monthly_expenses}! Now that we figured out your monthly expenses let's find out your monthly cash-flow!")
        self.cashFlow(monthly_expenses)
    
    
    def cashFlow(self, monthly_expenses):
        
        self.cash_flow = self.monthly_income - self.monthly_expenses
        print(f"Your total monthly cash-flow is {self.cash_flow}")
        print('We will find your total annual income next.')
        self.totalAnnualIncome()
    
    def totalAnnualIncome(self):
        self.annual_income = self.monthly_income * 12
        print(f"Your total annual income is {self.annual_income}. ")
        print('One more step before we find out your ROI.')
        self.totalInvestment()
        
    def totalInvestment(self):
        
        
        self.closing_costs = int(input("What were your closing costs? "))
        self.renovations = int(input("What was your downpayment? "))
        if self.house_value > 0:
            self.total_investment = self.house_value + self.closing_costs + self.renovations
        else:
            self.downpayment = int(input("What was your downpayment? "))
            self.total_investment = self.downpayment + self.closing_costs + self.renovations
            print(self.total_investment)
        self.roiCalculator()
        
    def roiCalculator(self):
        self.roi = (self.annual_income*100)/self.total_investment
        return print(f'Your ROI is {self.roi} %')
    

     
                
                
                      
        
testing = homeownerCashFlow()


        
```
